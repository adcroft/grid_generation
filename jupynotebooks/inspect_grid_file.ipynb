```python
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import numpy as np
%matplotlib inline
#Constants
PI_180 = np.pi/180.
#_default_Re = 6.378e6
_default_Re = 6371.e3 #MIDAS

def read_grid_file(fnam,format='NETCDF3_CLASSIC'):
    import netCDF4 as nc
    fin=nc.Dataset(fnam,'r',format=format)
    x=fin.variables["x"][:,:]
    y=fin.variables["y"][:,:]
    dx=fin.variables["dx"][:,:]
    dy=fin.variables["dy"][:,:]
    area=fin.variables["area"][:,:]
    angle_dx=fin.variables["angle_dx"][:,:]
    
    return x,y,dx,dy,area,angle_dx

def read_mask_file(fnam,format='NETCDF3_CLASSIC'):
    import netCDF4 as nc
    fin=nc.Dataset(fnam,'r',format=format)
    mask=fin.variables["mask"][:,:]
    return mask

```
