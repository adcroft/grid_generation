```python
import matplotlib.pyplot as plt
import netCDF4
import numpy
%matplotlib inline
```

```python
Orig05 = netCDF4.Dataset('/archive/gold/datasets/OM4_05/mosaic_ocean.v20180227.unpacked/ocean_hgrid.nc')
New05 = netCDF4.Dataset('ocean_hgrid_OM4p5.nc')
New05
```

```python
# xl=(350,370); yl=(1130,None);
# xl=(1410,None); yl=(1130,None);
# xl=(None,30); yl=(None,30);
xl=(None,None); yl=(None,None);
```

```python
plt.figure(figsize=(14,10)); 
plt.subplot(231); plt.pcolormesh( Orig05['x'][:]); plt.colorbar();plt.title('$\lambda$'); plt.xlim(xl); plt.ylim(yl);
plt.subplot(234); plt.pcolormesh( Orig05['y'][:]); plt.colorbar();plt.title('$\phi$'); plt.xlim(xl); plt.ylim(yl);
plt.subplot(232); plt.pcolormesh( Orig05['dx'][:]); plt.colorbar();plt.title('$\Delta x$'); plt.xlim(xl); plt.ylim(yl);
plt.subplot(235); plt.pcolormesh( Orig05['dy'][:]); plt.colorbar();plt.title('$\Delta y$'); plt.xlim(xl); plt.ylim(yl);
plt.subplot(233); plt.pcolormesh( Orig05['area'][:]); plt.colorbar();plt.title('A'); plt.xlim(xl); plt.ylim(yl);
plt.subplot(236); plt.pcolormesh( Orig05['angle_dx'][:]); plt.colorbar();plt.title('angle_dx'); plt.xlim(xl); plt.ylim(yl);
```

```python
plt.figure(figsize=(14,10)); 
plt.subplot(231); plt.pcolormesh( New05['x'][:]); plt.colorbar();plt.title('$\lambda$'); plt.xlim(xl); plt.ylim(yl);
plt.subplot(234); plt.pcolormesh( New05['y'][:]); plt.colorbar();plt.title('$\phi$'); plt.xlim(xl); plt.ylim(yl);
plt.subplot(232); plt.pcolormesh( New05['dx'][:]); plt.colorbar();plt.title('$\Delta x$'); plt.xlim(xl); plt.ylim(yl);
plt.subplot(235); plt.pcolormesh( New05['dy'][:]); plt.colorbar();plt.title('$\Delta y$'); plt.xlim(xl); plt.ylim(yl);
plt.subplot(233); plt.pcolormesh( New05['area'][:]); plt.colorbar();plt.title('A'); plt.xlim(xl); plt.ylim(yl);
plt.subplot(236); plt.pcolormesh( New05['angle_dx'][:]); plt.colorbar();plt.title('angle_dx'); plt.xlim(xl); plt.ylim(yl);
```

```python
plt.figure(figsize=(14,10)); 
plt.subplot(231); plt.pcolormesh( New05['x'][:] - Orig05['x'][:]); plt.colorbar();plt.title('$\lambda$'); plt.xlim(xl); plt.ylim(yl);
plt.subplot(234); plt.pcolormesh( New05['y'][:] - Orig05['y'][:]); plt.colorbar();plt.title('$\phi$'); plt.xlim(xl); plt.ylim(yl);
plt.subplot(232); plt.pcolormesh( New05['dx'][:] - Orig05['dx'][:]); plt.colorbar();plt.title('$\Delta x$'); plt.xlim(xl); plt.ylim(yl);
plt.subplot(235); plt.pcolormesh( New05['dy'][:] - Orig05['dy'][:]); plt.colorbar();plt.title('$\Delta y$'); plt.xlim(xl); plt.ylim(yl);
plt.subplot(233); plt.pcolormesh( New05['area'][:] - Orig05['area'][:]); plt.colorbar();plt.title('A'); plt.xlim(xl); plt.ylim(yl);
plt.subplot(236); plt.pcolormesh( New05['angle_dx'][:] - Orig05['angle_dx'][:]); plt.colorbar();plt.title('angle_dx'); plt.xlim(xl); plt.ylim(yl);
```

```python
# Y top edge
plt.figure(figsize=(14,5));
plt.subplot(121)
plt.plot( New05['y'][-1,:], label='new' );
plt.plot( Orig05['y'][-1,:], label='orig' );
plt.title('$\phi$ at j=ni');
plt.legend();
# Y symmetry
plt.subplot(122)
plt.plot( New05['y'][-2,:] - New05['y'][-2,::-1], label='new' );
plt.plot( Orig05['y'][-1,:] - Orig05['y'][-1,::-1], label='orig' );
plt.title('Symmetry test: $\phi$ at j=nj');
plt.legend();
```
